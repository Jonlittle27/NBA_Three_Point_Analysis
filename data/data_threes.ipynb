```python
#Dependencies
import pandas as pd
from sportsreference.nba.teams import Teams
```

```python
#Importing stats from the 2019-2020 NBA Season *A full regular season was not played due to Covid-19*
teams = Teams("2020")

stats_19_20 = teams.dataframes
stats_19_20.head()
```

```python
#subsetting for three point values
#
threes_20 = stats_19_20[['three_point_field_goal_attempts', 'three_point_field_goal_percentage',
       'three_point_field_goals', "games_played"]].copy()
threes_20
```

```python
#creating a new column for avg attempts/makes per game
threes_20["three_point_field_goal_attempts_per_game"] = round(threes_20["three_point_field_goal_attempts"] / threes_20["games_played"],2)
threes_20["three_point_field_goals_per_game"] = round(threes_20["three_point_field_goals"] / threes_20["games_played"],2)

threes_20
```

```python
#summary statistics for the 2019-2020 season for 3s
threes_20.describe()
```

```python
#nba data over the last 10 years
year = 2010
stats_10s = []
for i in range(10):
    year = str(year)
    new_year = Teams(year)
    new_year = new_year.dataframes
    new_year["year"] = year
    stats_10s.append(new_year)
    year = int(year)
    year += 1

stats_10s
```
